## Importing packages

```python
import slate3k as slate

from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

import glob
```

## PhD publications

### Importing the PDF files of publications

```python
publication_lists_phd = glob.glob('Publications_AN_PDFs_PhD\*\*\\*.pdf')
len(publication_lists_phd)
```

```python
abstract_phd = ''
for i in publication_lists_phd:
    with open(i,'rb') as f:
        extracted_text_phd = slate.PDF(f)
        abstract_phd += extracted_text_phd[0].split('Abstract')[1].split('Introduction')[0]
# print(abstract)
```

### creating word cloud

```python
wc_ri = WordCloud(background_color="white")
wc_ri.generate(abstract_phd)

plt.figure(figsize = (15,15))
plt.imshow(wc_ri, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### removing most frequent words

```python
stopwords = set(STOPWORDS)
stopwords.update(["used","type","is","are","based on","to reduce","results","is used", "is developed","to design",
                 "using", "developed", "to "])

wc_ri_wf = WordCloud(stopwords=stopwords, background_color="white", max_font_size=50, max_words=100).generate(abstract_phd)

plt.figure(figsize = (15,15))
plt.imshow(wc_ri_wf, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### put custom color

```python
custom_mask = np.array(Image.open("My_papers_mask.png"))
wc_f = WordCloud(stopwords=stopwords, background_color="black", mask=custom_mask)
wc_f.generate(abstract_phd)
image_colors = ImageColorGenerator(custom_mask)
wc_f.recolor(color_func=image_colors)

plt.figure(figsize = (15,15))
plt.imshow(wc_f, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### saving figure

```python
wc_f.to_file('AN_research_intrests_PhD.png')
```
